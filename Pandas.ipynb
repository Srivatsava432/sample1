# Pandas

- Built on top of the Python programming language, Pandas is a powerful library that, with the help of high-level data structures, makes the data wrangling and analysis process simple, easy and fast.

- Pandas main data structures are:
    * DataFrames
    * Series

- A Dataframe is a 2-dimensional data structure that organises data in rows and columns, similar to a spreadsheet. It contains variables of different data types.

- A Series is a one-dimensional array that has data of different types. Each value is associated with an index.


## Series

- Pandas series is capable of holding different data types

- Similar to a column in a spreadsheet

### Series Indexing

- Indexing is used to retrieve some or all elements of an array. Indexing is applied through:
    * index
    * loc
        - Retrieves data based on labels of rows and columns
    * iloc
        - Iloc function retrieves data based on index


```python
arr = np.array(["Audi", "BMW", "Mercedes", "Jaguar", "Toyota", "Honda", "RangeRover", "VW"])
cars = pd.Series(arr)

#Getting index
print("Slicing through index")
print(cars[0:3])
print("_"*20)
#loc
#cars.loc[0:3]
print("Slicing through loc")
print(cars.loc[0:3])
print("_"*20)
#iloc
#cars.iloc[0:3]
print("Slicing through iloc")
print(cars.iloc[0:3])
print(cars.iloc[0:3])
```

### Working with missing data

- Missing data can occur when no information is provided

- In Pandas, missing data is represented by two values: None and NaN (an acronym for Not a Number)

- None and NaN are essentially interchangeable for indicating missing or null values

- There are several useful functions for detecting, removing, and replacing null values in Pandas DataFrame

- These functions can also be used in Pandas Series:
    - isnull()
    - notnull()
    - dropna()
    - fillna()
    - replace()
    - interpolate()

### values_count()

```python
# How many times each value appear in a column/feature
df["Satisfaction"].value_counts()
```

```python
# How many times each item appear in the list
names = ["Robin", "Mat", "Sally", "Ahmed", "Rebecca", "Naveen", "Sara", "Robin", "Robin", "Sara", "Mat", "jude"]


name_count = {}
for name in names:
    if name in name_count:
        name_count[name] += 1
    else:
        name_count[name] = 1
for name, count in name_count.items():
    print(f"{name}: {count}")
```

```python
# We can use values_count instead
df = pd.DataFrame(names)
df[0].value_counts()
```

### Merge()

```python
df1 = pd.DataFrame([[121, "Mark", "Wood"], [122, "Ahmed", "Raza"], [123, "Sara", "Hadley"], [124, "Richard", "Savage"]],
     columns=['empID', 'First Name', "Last Name"])
df1
```

```python
df2 = pd.DataFrame([[124, "London", "Sales"], [122, "Paris", "IT"], [121, "Barcelona", "Sales"], [123, "Budapest", "Marketing"]],
     columns=['ID', 'City', "Department"])
df2
```

```python
df1.merge(df2, how= "left",left_on='empID', right_on="ID")
```

Also see how="right" and "inner"

### Pivot()

```python
# create a pivot table
df2 = df2.pivot(index='ID', columns='Department', values='City')
df2
```
